**pre-processing an image**

```python
file = "/content/Bird-Cock.jpg"
```

```python
from tensorflow.keras.preprocessing import image
img = image.load_img(file, target_size=(224,224))
```

```python
plt.imshow(img)
```

```python
resized_image = image.array_to_img(img)      # (224, 224, 3)
final_img = np.expand_dims(resized_image, axis=0)    ## need fourth dimenssion
final_img = tf.keras.applications.mobilenet.preprocess_input(final_img)
```

```python
final_img.shape
```

```python
prediction = mobile.predict(final_img)
```

```python
prediction      # we dont know what it is, because we do not provide them what are the label, so to check your ans true or not
```

```python
from tensorflow.keras.applications import imagenet_utils
```

```python
results = imagenet_utils.decode_predictions(prediction)
```

```python
print(results)
```
